#Python scope
A variable is only available from inside  the region is created. This is called scope

#**Local scope**
Created inside the function and it can only be used inside that function.








```python
def myFunc():
  i = 21
  print(i)
myFunc()
```

#**Function inside the function**

Local variable can be used for any function inside the function but cannot be used outside







```python
def myfunc():
  x = 21
  def innerFunc():
    print(x)
  innerFunc()
myFunc()
```

#**Global scope**

A variable created inside the body of the python code and can be used inside and outside of a function.















```python
x = 27
def myFunc():
  print(x)

myFunc()
print(x)
```

#**Naming variable**
if you create a variabe with same name inside and outside of a function Python will treat them as two seperate variables - one available in the global scope(outside the function) and other in the local scope(inside the function)












```python
x = 300
def myfunc():
  x = 200
  print(x)
myfunc()

print(x)
```

#**Global Keyword**
if yo need to created a global variable inside a function then you can use **global** keyword

**if you want to change the global variable inside a function you can use "global" keyword**








```python
def myFunc():
  global i
  i = 27
  print(i)
myFunc()
print(i)


x = 21
def myfunc():
  global x
  x = 27
  print(x)
myfunc()
print(x)
```

#nonlocal Keyword

The **nonlocal** keyword is used to work with variable inside nested functions
**nonlocal** keyword makes the variable belongs to the outer function







```python
def myfunc1():
  x = 21
  def myfunc2():
    nonlocal x
    x = 27
    print(x)
  myfunc2()
print(x)
myfunc1()
```

```python
def TwoSum(arr, target):
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[j] + arr[i] == target:
                return i,j
    return False
arr = [2, 3, 7, 5, 8]
target = 8
print(TwoSum(arr, target))
```
